```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
n = 1024
# 创造均值为0，方差为1的随机数；总共生成n个
X = np.random.normal(0, 1, n)
Y = np.random.normal(0, 1, n)
```

```python
# for color value
# 这一步只是为了好看
T = np.arctan2(Y, X)
```

```python
# s代表size；c代表color,T表示color map；alpha表示透明度
plt.scatter(X, Y, s=75, c=T, alpha=0.5)
```

```python
plt.xlim((-1.5, 1.5))
```

```python
plt.ylim((-1.5, 1.5))
```

```python
# 在xticks和yticks里面什么都不放，坐标的数值就会消失
plt.xticks(())
plt.yticks(())
```
