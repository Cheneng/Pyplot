```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(-1, 1, 500)
```

```python
y1 = 2 * x + 1
```

```python
y2 = x ** 2
```

```python
# 参数num代表是fugure几，而figsize用来代表尺寸（并不是axis，而是图片的大小）
plt.figure(num=3, figsize=(10, 5))
```

```python
plt.plot(x, y1)
```

```python
plt.figure()
```

```python
plt.plot(x, y2)
```

```python
# 使用figure来让图像显示在不同方框中去
plt.show()
```

```python
plt.figure(figsize=(10, 5))
# 这里调用了fugure，那么下面的画图东西都是跟figure有关的（即都会在同一张图片中去）
```

```python
# 使用两个plot，即在同一个figure中画两个图片
plt.plot(x, y2)

# color参数来表示画这个线的颜色，linewidth表示线的宽度， 
#linestyle表示这个线的风格，这里使用的是虚线--
plt.plot(x, y1, color='red', linewidth=1.0, linestyle='--')
plt.show()
```
